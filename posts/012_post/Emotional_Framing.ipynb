## Introduction

The way information is framed emotionally can significantly influence how it is interpreted, even when the underlying facts remain the same. In this post, I explore how an LLM adjusts tone and emotional framing when asked to present identical information with different emotional goals.

**Objective**
To test whether Perplexity can maintain factual consistency while altering emotional tone, and to observe how framing shapes perceived meaning.

## Prompts

1. Describe climate change in a neutral tone

3. Describe climate change in a tone that inspires urgnecy

4. Describe climate change in a reassuring tone

## Responses

1. Climate change refers to long-term shifts in global or regional climate patterns, particularly those that have become more apparent since the late 20th century. It involves changes in temperature, precipitation, wind patterns, and other aspects of the Earth’s climate system.

These changes can result from both natural processes, such as variations in solar radiation or volcanic activity, and human activities, including the burning of fossil fuels, deforestation, and industrial emissions. Scientists measure and model these factors to understand how the atmosphere, oceans, and land interact over time.

While there is broad scientific agreement that the Earth’s climate is warming, discussions about climate change often include different perspectives on its causes, impacts, and the best ways to address it. The topic intersects with environmental, economic, and social considerations, making it a complex global issue that requires coordinated research and policy responses.

2. Climate change is not a distant threat — it’s happening now, and its pace is accelerating. Rising temperatures are fueling extreme weather events, melting polar ice, and driving sea levels higher, threatening coastal communities and ecosystems that have existed for millennia. Wildfires burn larger areas, droughts last longer, and floods strike harder, disrupting food systems and endangering millions of lives.

Every fraction of a degree matters. The choices made today — to reduce emissions, protect forests, and transition to clean energy — will determine whether future generations inherit a planet that is livable or one defined by crisis and loss. The window to act is narrowing fast, but change remains possible. Human ingenuity and collective will can still shape a safer, more resilient world if action happens now.

## Reflection

It is very interesting how the framing altered some does not alter the factual elements of the responses, rather just the tone of urgency towards the subject. While slightly in between fact and opinion, the topic of climate change is one that is highly contested. In this example, Perplexity presented the same facts, but altering the way they were phrased and what was emphasized. In both the prompt asking for a reassuring tone and the prompt requesting a tone that inspires urgency, Perplexity awknoledges that climate change is an important and serious threat, but shifts the framing in how urgent or worried people should be, without altering the factual information it is presenting.
