One of the most useful ideas in A Brief History of Intelligence is Max Bennett’s argument that intelligence did not emerge all at once, but rather through a series of distinct evolutionary breakthroughs. Early in the book, Bennett outlines how capabilities like memory, learning, abstraction, and planning appeared incrementally across different species, each building on the last. This framework immediately reframed how I think about intelligence — not as a single measurable trait, but as a layered system of abilities that evolved to solve specific survival problems.

What stood out to me most is how this perspective challenges the way we talk about artificial intelligence today. We often ask whether AI is “intelligent” in a broad, almost binary sense. Bennett’s framework suggests this question may be misguided. Instead of asking whether AI is intelligent, we should ask which components of intelligence it replicates and which it does not. Large language models, for example, demonstrate impressive abstraction and pattern recognition, but they lack many of the adaptive and goal-oriented features that characterize biological intelligence.

This made me reconsider the tendency to compare AI directly to humans. Bennett’s evolutionary lens suggests that modern AI systems might resemble earlier stages of intelligence rather than a complete cognitive system. Understanding intelligence as cumulative rather than absolute helps explain why AI can appear brilliant in some contexts while failing spectacularly in others.

Ultimately, this section of the book taught me that intelligence is not a single finish line. It is a process — one shaped by environment, necessity, and gradual innovation — and AI occupies only a narrow slice of that broader evolutionary story.
