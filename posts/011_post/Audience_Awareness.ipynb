## Blog Task Overview

**What is this blog post doing?**

This post tests whether an LLM can recognize audience differences and adapt its communication style accordingly, a core concept in communication studies.

In human communication, effective messaging depends on:

Who the audience is

What prior knowledge they have

What tone and level of detail is appropriate

This blog evaluates whether LLMs can perform this same kind of audience-aware framing, or whether they simply simplify language without truly adapting meaning.

**Objective**
The task in this experiment is to ask the LLM to explain the same concept to three different audiences: a middle school student, a college professor, and a business executive. By holding the topic constant and changing only the audience, this test isolates whether the model can meaningfully adjust its communication style.

## Prompt
Explain how artificial intelligence works to:
1. A middle school student
2. A college professor
3. A business executive

## Reflection

Altough the LLM definitely did create different leveled responses for each age group, it was interesting that the length and structure of response remained the same. All three were bullet pointed and in short succinct sentences to describe the topic in just as much depth, but in different terms. I have noticed this is GPT's "go to" approach when responding to different questions and requests, usually responding with a brief overview and subsequent bullet points outlining the answer.
