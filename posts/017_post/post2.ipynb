**Overview**
This post reflects on Bennett’s explanation of why intelligence evolved in the first place and why it was neither inevitable nor universal.

One of the earliest ideas that stood out to me in A Brief History of Intelligence is Bennett’s argument that intelligence is not a default feature of life. Early in the book, he emphasizes that intelligence evolved because it solved specific survival problems, not because evolution naturally trends toward complexity or reasoning. Many organisms thrive without anything resembling human cognition, which immediately challenges the assumption that intelligence is the pinnacle of evolution.

What I found especially compelling is Bennett’s insistence that intelligence is costly. Brains require energy, coordination, and time to develop, meaning that intelligence only emerges when its benefits outweigh its risks. This framing helped me see intelligence not as a universal goal, but as a strategic adaptation shaped by environment and necessity.

Starting the book this way reframed how I think about both humans and AI. If intelligence is situational rather than inevitable, then it makes less sense to treat artificial intelligence as the natural “next step” in evolution. Instead, it becomes something that emerged under very specific conditions — biological ones that AI does not share.
