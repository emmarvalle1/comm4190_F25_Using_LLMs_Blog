As the book progresses, Bennett introduces one of its central frameworks: the idea that intelligence evolved through a sequence of major breakthroughs rather than appearing fully formed. Abilities such as memory, learning, and abstraction emerged at different points in evolutionary history, each enabling new ways of interacting with the world. This breakdown made intelligence feel more concrete and less mystical.

What struck me most is how this framework destabilizes the way we usually talk about intelligence as something you either have or don’t have. Instead, intelligence becomes modular — composed of different capacities that can exist independently or in partial forms. This perspective makes it easier to understand why some animals excel in certain cognitive tasks while lacking others.

Applying this framework to modern AI also felt intuitive. Rather than asking whether AI is intelligent in general, Bennett’s model encourages us to ask which breakthroughs AI mirrors and which it does not. This shift in thinking helped me see AI not as an imitation of humans, but as a system that reflects only certain layers of intelligence.
