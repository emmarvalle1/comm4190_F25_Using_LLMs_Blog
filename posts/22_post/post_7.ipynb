Near the end of A Brief History of Intelligence, Bennett challenges the way we frame debates about artificial intelligence. Rather than asking whether AI is intelligent in the human sense, he suggests that this question obscures more than it reveals. Intelligence, as the book repeatedly shows, is not a single trait but a collection of capacities shaped by evolutionary pressures.

What I found most useful about this reframing is how it lowers the emotional stakes of AI discourse. Bennett does not argue that AI is either a threat or a miracle. Instead, he encourages readers to evaluate AI systems based on what they can do, what they cannot do, and how those abilities fit within a much longer history of cognitive development.

This perspective helped me move away from comparisons between AI and humans and toward a more functional understanding of intelligence. By treating AI as a specific configuration of cognitive tools rather than a rival mind, Bennett offers a way to discuss its impact without resorting to exaggeration or fear.
