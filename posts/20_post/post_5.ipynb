**Description:** This post considers Bennett’s distinction between prediction and understanding and what it reveals about the capabilities and limits of modern AI.

In the later sections of the book, Bennett draws a clear distinction between prediction and understanding — a distinction that feels especially relevant to large language models. While AI systems are exceptionally good at predicting likely outcomes or word sequences, Bennett argues that understanding requires internal models, goals, and the ability to adapt based on consequences.

This idea helped me reinterpret both the strengths and failures of AI. When a language model produces a confident but incorrect response, it is not “misunderstanding” in a human sense — it is still doing exactly what it was designed to do. The mistake lies in assuming that fluent prediction implies comprehension.

By the end of the book, Bennett applies this framework to modern AI without either exaggerating or dismissing its significance. AI appears powerful because it replicates certain cognitive functions, but it lacks the deeper structures that make biological intelligence adaptive and self-directed. This conclusion left me with a more grounded way of thinking about AI — not as a replacement for human intelligence, but as a tool that occupies a specific place within a much longer history of cognition.
