Bennett repeatedly emphasizes that intelligence cannot be fully understood without understanding evolution. Rather than treating cognition as an abstract problem-solving tool, the book frames intelligence as something shaped by millions of years of trial and error. Brains evolved not to discover truth, but to improve survival under specific environmental constraints.

What stood out to me is how this evolutionary framing challenges the idea that intelligence is neutral or objective. Human reasoning is deeply shaped by biases, heuristics, and shortcuts because those traits were often advantageous, even if they were imperfect. Intelligence, from this perspective, is practical rather than ideal.

This insight carries important implications for how we think about AI. Because AI does not evolve through natural selection, it does not inherit the same priorities, blind spots, or adaptive pressures. While this can make AI more precise in some contexts, it also means it lacks the intuitive grounding that evolution provided to biological intelligence.
